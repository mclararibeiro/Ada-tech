```python
import pandas as pd

dados = pd.read_csv('churn.csv')
display(dados)
```

```python
dados.info()
```

```python
#Porcentagem de 'Churn' na tabela
print(dados['Exited'].value_counts())
print(dados['Exited'].value_counts(normalize=True))
```

```python
#Trocando o Ã­ndice para a primeira coluna
tabela = dados.set_index('RowNumber')
tabela
```

```python
tabela['NumOfProducts'].nunique()
```

```python
tabela_churn = tabela.query('Exited ==  1')
tabela_churn
```

```python
tabela_churn.describe().round(0)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize =(8,5))
plt.title("AnÃ¡lise por PaÃ­s")
plt.bar(tabela_churn['Geography'].value_counts().index,tabela_churn['Geography'].value_counts());
plt.xticks(rotation = 45, ha ='right');
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize =(8,5))
plt.title("AnÃ¡lise por Genero")
plt.bar(tabela_churn['Gender'].value_counts().index,tabela_churn['Gender'].value_counts());
plt.xticks(rotation = 45, ha ='right');
```

```python
dados_idade = tabela_churn['Age'].value_counts()
dados_idade.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize =(8,5))
plt.title("AnÃ¡lise por idade")
plt.stem(dados_idade.index,dados_idade);
plt.xlabel('Idade')
plt.ylabel('Quantidade de perda de clientes')
plt.xticks(rotation = 45, ha ='right');
```

```python
dados_produtos = tabela_churn['NumOfProducts'].value_counts()
dados_produtos 
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize =(8,5))
plt.title("AnÃ¡lise por Assinaturas de Produtos")
plt.bar(dados_produtos.index,dados_produtos, color = '#a1d99b');
plt.xlabel('Produtos assinados')
plt.ylabel('NÃºmero de assinantes')
plt.xticks(rotation = 45, ha ='right');
```

```python
confiabilidade = tabela_churn['Tenure'].value_counts()
confiabilidade
```

```python
plt.figure( figsize=(15,6))
plt.bar(confiabilidade.index, confiabilidade)
plt.title('AnÃ¡lise da estabilidade vs churn')
plt.ylabel("Quantidade de clientes desligados")
plt.xticks( rotation=45, ha='right');
```

```python
#Agrupando clientes desligados com estabilidade acima de 7
estabilidade_alta = tabela_churn.loc[tabela_churn['Tenure'] >= 7]
estabilidade_alta
```

```python
#Quantidade de membros com estabilidade alta e desligados que eram membros ativos
membros_ativos = estabilidade_alta['IsActiveMember'].value_counts()
membros_ativos
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize =(15,6))
plt.title("Membros ativos - Clientes desligados com estabilidade alta")
plt.pie(membros_ativos, labels = membros_ativos.index, startangle=90, autopct='%1.1f%%')
plt.show()
```

```python
#Membros ativos churn
membros_ativos_totais = tabela_churn['IsActiveMember'].value_counts(normalize=True)
membros_ativos_totais
```

```python
#Agrupando clientes por creditscore
creditscore = tabela_churn.loc[tabela_churn['CreditScore'] > 700]
creditscore
```
